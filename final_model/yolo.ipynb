```python
from ultralytics import YOLO

# Load a YOLOv8 classification model
model = YOLO("yolov8n-cls.pt")  # small model, faster training; use yolov8s-cls.pt for bigger

# Train
results=model.train(
    data="D:/ML/Plant_classification_proj/data/30_plant_species_split",  # folder with subfolders = class names
    epochs=50,
    imgsz=224,                # image size
    batch=32,                 # adjust if you get OOM errors
    device=0,                 # GPU 0
)
```
