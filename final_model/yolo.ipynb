```python
from ultralytics import YOLO


model = YOLO("yolo11s-cls.pt") 

results = model.train(
    data="D:/ML/Plant_classification_proj/data/25_plant_species_split",
    epochs=100,
    imgsz=320,
    batch=32,
    device=0,
    patience=10,
    augment=True,  
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=15,
    translate=0.1,
    scale=0.5,
    shear=2.0,
    fliplr=0.5,
    mixup=0.2,
)
```

```python
import os


dataset_dir = "../data/25_plant_species_split/train"  

classes = [d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))]

class_counts = {}
for cls in classes:
    cls_path = os.path.join(dataset_dir, cls)
    class_counts[cls] = len([f for f in os.listdir(cls_path) if os.path.isfile(os.path.join(cls_path, f))])

print("Image counts per class:")
for cls, count in class_counts.items():
    print(f"{cls}: {count}")

print("\n Total images in dataset:", sum(class_counts.values()))

```
